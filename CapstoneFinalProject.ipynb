```python
import pandas as pd
import numpy as np
import requests

from bs4 import BeautifulSoup


source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text

soup = BeautifulSoup(source, 'html5lib')

postal_codes_dict = {} # initialize an empty dictionary to save the data in



```

```python
postalcode = []
borough = []
neighbourhood = []
for i, row in enumerate(soup.find_all("td")):
    text = row.text[:-1]
    if i % 3 == 0:
        postalcode.append(text)
    if i % 3 == 1:
        borough.append(text)
    if i % 3 == 2:
        if text == 'Not assigned' and i >1:
            neighbourhood.append(str(borough[-1]))
        else:
            neighbourhood.append(text)

  
neighbourhood = neighbourhood[:180]
borough = borough[:180]
postalcode = postalcode[:180]

toronto_data = pd.DataFrame()
toronto_data['Postal Code'] = postalcode
toronto_data['Borough'] = borough
toronto_data['Neighborhood'] = neighbourhood
toronto_data
```

```python
toronto_data = toronto_data[toronto_data.Borough!='Not assigned']
toronto_data=toronto_data.reset_index()
toronto_data.drop(['index'], axis=1,inplace=True)
toronto_data
```
